<p style="font-size:17px">
Problem statement: Find relation between "species" and "petal_lenght"<br>
H0(Null Hypothesis): "Species" and "petal_lenght" are independent<br>
H1(Alternative Hypothesis): "Species" and "petal_lenght" are dependent<br>

```python
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
*speciesGrouped, = df.groupby('species')
```

```python
len(speciesGrouped)
```

```python
setosa = speciesGrouped[0][1]['petal_length']
versicolor = speciesGrouped[1][1]['petal_length']
virginica = speciesGrouped[2][1]['petal_length']
```

```python
f_oneway(setosa,versicolor,virginica)
```

```python
# Alternate method for anova test
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('petal_length ~ species', data = df).fit()
sm.stats.anova_lm(model, typ=2)
```

<p style="font-size:17px">
pvalue < 0.05<br>
Null Hypothesis(H0) is Rejected<br>
The "Species" and "Petal Length" are dependent<br>
